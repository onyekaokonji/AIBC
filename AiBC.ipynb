```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from keras.models import Input
from keras.layers import Dropout
from tensorflow.keras.optimizers import *
```

```python
# Import the dataset
train_data = pd.read_csv('/content/Train.csv')
test_data = pd.read_csv('/content/Test.csv')
```

## exploratory data analysis 

```python
train_data.head(11)
```

```python
train_data['Response'].value_counts()
```

```python
Response_0, Response_1 = train_data['Response'].value_counts()
```

```python
train_data_0 = train_data[train_data['Response'] == 0]
train_data_1 = train_data[train_data['Response'] == 1]
```

```python
train_data_1_over = train_data_1.sample(Response_0, replace = True)
new_train_data = pd.concat([train_data_0, train_data_1_over], axis = 0)
```

```python
print(new_train_data['Education_Level'].unique())
print(new_train_data['Marital_Status'].unique())
```

```python
print(test_data['Education_Level'].unique())
print(test_data['Marital_Status'].unique())
```

```python
# We drop Marital Status because it isn't equal across datasets and Customer Date isn't relevant to accuracy
new_train_data.drop(['Marital_Status', 'Date_Customer'], axis = 1, inplace= True)
```

```python
test_data.drop(['Marital_Status', 'Date_Customer'], axis = 1, inplace=True)
```

```python
# Check for missing values in training data
new_train_data.isnull().sum()
```

```python
# Check for missing values in test data
test_data.isnull().sum()
```

```python
# Handling missing data

new_train_data.fillna(method='ffill', inplace=True)
test_data.fillna(method='ffill', inplace=True)
```

```python
new_train_data.isnull().sum()
```

```python
# Encoding categorical data
labelencoder = LabelEncoder()
new_train_data['Education_Level'] = labelencoder.fit_transform(new_train_data['Education_Level'])
```

```python
test_data['Education_Level'] = labelencoder.fit_transform(test_data['Education_Level'])
```

```python
test_data
```

```python
#  Extracting features and target data
train_data_x = new_train_data.iloc[:, 0:24]
train_data_y = new_train_data.iloc[:, 24]
```

```python
# Scaling the features to a range
s_scaler = StandardScaler()
train_data_x.iloc[:, 1:24] = s_scaler.fit_transform(train_data_x.iloc[:, 1:24])
test_data.iloc[:, 1:24] = s_scaler.fit_transform(test_data.iloc[:, 1:24])
```

```python
# Splitting train data to obtain validation_data
x_train, x_val, y_train, y_val = train_test_split(train_data_x.iloc[:, 1:24], train_data_y, test_size= 0.2, shuffle=True)
```

```python
x_train = np.asarray(x_train).astype('int')
x_val = np.asarray(x_val).astype('int')
y_train = np.asarray(y_train).astype('int')
y_val = np.asarray(y_val).astype('int')
```

# model architecture and training

```python
sapa = Sequential()
sapa.add(Input(shape=(23,)))
sapa.add(Dense(units=32, activation='relu', kernel_initializer='RandomUniform'))
sapa.add(Dense(units= 18, activation='relu', kernel_initializer='RandomUniform'))
sapa.add(Dropout(0.2))
sapa.add(Dense(units= 18, activation='relu', kernel_initializer='RandomUniform'))
sapa.add(Dense(units= 14, activation='relu', kernel_initializer='RandomUniform'))
sapa.add(Dropout(0.1))
sapa.add(Dense(units= 14, activation='relu', kernel_initializer='RandomUniform'))
sapa.add(Dense(units= 10, activation='relu', kernel_initializer='RandomUniform'))
sapa.add(Dense(units= 1, activation='sigmoid'))
```

```python
optim = Adam(learning_rate=0.0001, beta_1=1.2, )
sapa.compile(optimizer=optim, loss='binary_crossentropy', metrics=['Accuracy'])
check = sapa.fit(x = x_train, y = y_train, batch_size=8, epochs=200, validation_data=(x_val, y_val))
```

```python
pred = sapa.predict(test_data.iloc[:, 1:24])
```

```python
pred = np.rint(pred)
```

```python
pred = pred.reshape(-1)
```

# generating submission file
